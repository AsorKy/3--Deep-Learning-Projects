#3)-Data Exploration

We can separate the three color channels of an image just by selecting the last dimension of it.

```python
plt.imshow(im)
```

```python
im_r = im[:,:,0]
im_g = im[:,:,1]
im_b = im[:,:,2]
```

When we try to plot just one channel of the image, python automatically plots the gray scale of just that channel.

```python
plt.imshow(im_g)
```

In order to see the layer of each RGB component, we need to stack null values in each of the channels that we do not want to see, in order to do so, we create auxiliary dimensions of color as matrices filled of zerons and tend we stack them in a single tensor to subsequently plot the image as a three channel image.

```python
# Auxiliary dimensions
aux_d = im_r.shape
aux_dim = np.zeros([aux_d[0],aux_d[1]])

# Dimensionality stack
RGB_Channels = []
RGB_Channels.append( np.dstack((im[:,:,0], aux_dim, aux_dim)).astype(np.uint8))
RGB_Channels.append( np.dstack((aux_dim, im[:,:,1], aux_dim)).astype(np.uint8))
RGB_Channels.append( np.dstack((aux_dim, aux_dim, im[:,:,2])).astype(np.uint8))

# Image Plot
i=1
plt.imshow(RGB_Channels[i])
plt.show()
```

We can also create a neggative image just by taking the maximum pixel value and then substacting the actual value of the pixel in the original image.

```python
im_neg_pos = 255 - im
plt.imshow(im_neg_pos)
```

Treating images as matrices allows us to manipulate them very easilly, for example, we could reduce the image reasulution by pixel quantization, or, we culd make a zoom to the image

```python
im32 = (im // 32)*32
im128 = (im // 128)*128
plt.imshow(im32)
```
