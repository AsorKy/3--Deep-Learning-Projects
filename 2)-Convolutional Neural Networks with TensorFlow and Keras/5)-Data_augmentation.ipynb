#Data Augmentation

#1)-Library Importing

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import matplotlib.pyplot as plt
from skimage import io
```

##2)-Data Importing

```python
path = '/content/drive/MyDrive/Colab Notebooks/2)-Machine Learning/2)-Platzy Career (Deep Learning)/2)-Convolutional Neural Networks with TensorFlow and Keras/data/vishnu_1.jpeg'
im = io.imread(path)
print(im.shape)
im = im.reshape((1,)+ im.shape)
print(im.shape)
```

#3)-Data Augmentation

```python
datagen = ImageDataGenerator(rotation_range=40,
                   width_shift_range=0.2,
                   height_shift_range=0.2,
                   zoom_range=0.2,
                   horizontal_flip=True,
                   fill_mode='nearest',
                   brightness_range=[0.4,1.5])
```

```python
i = 0
for batch in datagen.flow(im, batch_size=1):
  plt.figure(i)
  implot = plt.imshow(array_to_img(batch[0]))
  i += 1
  if i % 10 == 0:
    break
  plt.show()
```

```python
train_generator = datagen.flow_from_directory('/content/drive/MyDrive/Colab Notebooks/2)-Machine Learning/2)-Platzy Career (Deep Learning)/2)-Convolutional Neural Networks with TensorFlow and Keras/data/cats_and_dogs/train',
                                             target_size=(150,150),
                                             batch_size=32,
                                             class_mode='binary')
```
