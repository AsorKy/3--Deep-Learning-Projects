#Kernel manipulation

#1)-Importing Libraries


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as nd
from skimage import io, color
```

#2)-Data Importing

```python
path = '/content/drive/MyDrive/Colab Notebooks/2)-Machine Learning/2)-Platzy Career (Deep Learning)/2)-Convolutional Neural Networks with TensorFlow and Keras/data/CV Image.jpg'
img = io.imread(path)
```

#3)-Data Exploration

```python
img_gray = color.rgb2gray(img)
plt.imshow(img_gray)
plt.show()
```

```python
img_gray.shape
```

#4)-Image Manipulation via stipulated kernels

```python
kernel_vertical = np.array([[-1,0,1],
                            [-1,0,1],
                            [-1,0,1]])
kernel_horizontal = np.array([[-1,-1,-1],
                              [0,0,0],
                              [1,1,1]])
```

```python
img_new_vert = nd.convolve(img_gray, kernel_vertical)
img_new_hori = nd.convolve(img_gray, kernel_horizontal)

fig, axes = plt.subplots(2,2,figsize=(15,10))
axes[0,0].imshow(img_gray, cmap=plt.cm.gray)
axes[0,0].set_title('Original Image')
axes[0,0].axis('off')
axes[0,1].imshow(img_new_vert, cmap=plt.cm.gray)
axes[0,1].set_title('Conv. Vertical Image')
axes[0,1].axis('off')
axes[1,0].imshow(img_gray, cmap=plt.cm.gray)
axes[1,0].set_title('Original Image')
axes[1,0].axis('off')
axes[1,1].imshow(img_new_hori, cmap=plt.cm.gray)
axes[1,1].set_title('Conv. Horizontal Image')
axes[1,1].axis('off')
plt.show()
```

```python
imgplus = img_new_vert + img_new_hori
plt.imshow(imgplus, cmap=plt.cm.gray)
```
