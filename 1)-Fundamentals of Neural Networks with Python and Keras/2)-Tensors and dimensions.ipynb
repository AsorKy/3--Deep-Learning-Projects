```python
import numpy as np
```

### Scalar

```python
x = np.array(42)
print(x.shape)
x.ndim
```

## Vector

```python
x = np.array([42,23,453,21,32,1,3,7])
print(x.shape)
x.ndim
```

## Matrix

```python
x = np.array([[42,23,453],
             [34,343,24]])
print(x.shape)
print(x.size)
x.ndim
```

## Tensor

```python
x = np.array([[[42,23,453],
             [34,343,24]],
             [[42,23,453],
             [34,343,24]],
             [[42,23,453],
             [34,343,24]],
             [[42,23,453],
             [34,343,24]]])
print(x.shape)
print(x.size)
x.ndim
```

## Reshape

```python
x = np.array([[0,1],
             [2,3],
             [4,5],
             [6,7]])
print(x.size)
x.shape
```

```python
x.reshape(8,1)
```

```python
x.reshape(2,4)
```

```python
np.transpose(x)
```

```python
x.T
```
